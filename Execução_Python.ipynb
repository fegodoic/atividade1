```python
#importacao dos módulos
from mpl_toolkits import mplot3d
import pandas as pd
import matplotlib.pyplot as plt
import numpy as ny
```

```python
#declara as variaveis do grafico
valor_y = pd.read_table('Arq02.txt')
valor_x = pd.read_table('Arq02.txt')
valor_z = pd.read_table('resultados.txt')

#declara a figura
fig = plt.figure()

#declara as 3 dimensoes do grafico
grafico = plt.axes(projection = '3d')
grafico = plt.axes(projection = "3d")

#passa as variaveis e personalizacao
grafico.plot_wireframe(valor_x, valor_y, valor_z, color='purple')
grafico.set_xlabel('eixo X')
grafico.set_ylabel('eixo Y')
grafico.set_zlabel('eixo Z')
grafico.set_title('Gráfico');

#comando para plotar o grafico
plt.show()
```
